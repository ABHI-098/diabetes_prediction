```python
import numpy as np
import pandas as pd
df=pd.read_csv("pima-indians-diabetes.csv")
```

```python
dfmap={True: 1,False: 0}
df['diabetes']=df['diabetes'].map(dfmap)
df.head()
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,1]
```

```python
xtr,xte,ytr,yte=tts(x,y,test_size=0.1)
```

```python
for i in ['num_preg','glucose_conc','diastolic_bp','thickness','insulin','bmi','diab_preb','age']:
    print('no. of rows missing in {0}: {1}'.format(i,len(df.loc[df[i]==0])))
```

```python
from sklearn.preprocessing import Imputer
fill=Imputer(missing_values=0,strategy="mean",axis=0)
xtr=fill.fit_transform(xtr)
xte=fill.fit_transform(xte)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfm = RandomForestClassifier(random_state=10)

rfm.fit(xtr, ytr.ravel())
```

```python
predict_train_data = rfm.predict(xte)

from sklearn import metrics
```

```python
print("Accuracy =",(metrics.accuracy_score(yte, predict_train_data)))
```
